# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Passo a Passo para resolver o desafio 

Passo 1 - Entrar no sistema da empresa (entrar no link: https://drive.google.com/drive/folders/1dmyUobeAx2u6-csCwljkJI6gLCX1_nOq?usp=sharing)

Passo 2 - Navegar até o local da base de dados (Entrar na pasta "Exportar")

Passo 3 - Baixar a planilha de vendas

Passo 4 - Calcular o faturamento e quantidade de produtos vendidos da base

Passo 5 - Enviar email para a diretoria com a quantidade e o faturamento que calculamos

```python
import pyautogui
import time
import pyperclip

pyautogui.PAUSE = 1

# Primeiro Passo
# Abrir uma nova aba
pyautogui.hotkey('ctrl', 't')
# Entrar no link do sistema
link = "https://drive.google.com/drive/u/0/folders/1dmyUobeAx2u6-csCwljkJI6gLCX1_nOq"
pyperclip.copy(link)
pyautogui.hotkey("ctrl", "v")
pyautogui.press('enter')

# Segundo Passo
time.sleep(5)
pyautogui.click(446, 347, clicks=2)
time.sleep(2)

# Terceiro Passo
pyautogui.click(471, 532) # selecionar o arquivo
pyautogui.click(1657, 216) # clicar em "mais ações"
pyautogui.click(1441, 715) # clicar em "fazer download"
time.sleep(10)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Quarto Passo
import pandas as pd

tabela = pd.read_excel(r"C:\Users\Gabriel\Downloads\Vendas - Dez.xlsx")
faturamento = tabela["Valor Final"].sum() # soma da coluna Valor Final
quantidade = tabela["Quantidade"].sum() # soma da coluna Quantidade
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Quinto Passo

# Abrir uma nova aba
pyautogui.hotkey('ctrl', 't')
# Entrar no Gmail
link= "https://mail.google.com/"
pyperclip.copy(link)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")
time.sleep(5)
# Clicar no botão escrever
pyautogui.click(96, 196)
# Digitar o email para envio
pyautogui.write("gabrielpierrout@gmail.com")
pyautogui.press("tab") # escolher email
pyautogui.press("tab") # descer para campo de assunto
# Digitar o assunto
assunto = "Relatório de Vendas"
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") # descer para campo de digitar email

# Digitar o corpo do email
texto_email = f"""
Prezados, boa noite

O faturamento de ontem foi: R${faturamento:,.2f} 
A quantidade de produtos foi de: {quantidade:,}

Abs
Gabriel Pierrout
"""

pyperclip.copy(texto_email)
pyautogui.hotkey("ctrl", "v")

# Clicar em enviar
pyautogui.hotkey("ctrl", "enter")
pyautogui.alert("Operação Finalizada!")
```
